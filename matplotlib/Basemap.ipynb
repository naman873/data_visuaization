```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
m=Basemap(projection='mill',llcrnrlat=-90,llcrnrlon=-180,urcrnrlat=90,urcrnrlon=180,resolution='l')
```

```python
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='c')
m.drawcountries(color='darkred')
#m.fillcontinents()
m.bluemarble()
#m.etopo()
plt.title("Basemap")
plt.show()
```

```python
m=Basemap(resolution='l',projection='mill',llcrnrlat=25,llcrnrlon=-130,urcrnrlat=50,urcrnrlon=-60)
```

```python
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')

xs=[]
ys=[]

NYClat,NYClon=40.7127,-74.0059
xpt,ypt=m(NYClon,NYClat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt,ypt,"c*",markersize=15)

LAlat,LAlon=34.05,-118.25
xpt,ypt=m(LAlon,LAlat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt,ypt,'g^',markersize=15)

m.plot(xs,ys,linewidth=2,color='r',label='flight98')
m.drawgreatcircle(NYClon,NYClat,LAlon,LAlat,linewidth=3,color='c',label="cicle _ex")
m.bluemarble()
plt.legend(loc=4)  ### loc is defind by location we can add location from range 0 to 10 specifed in info of leg
plt.show()
```
