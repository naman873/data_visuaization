```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
x=np.array([1,2,3,7,4,5,6,7,5])
y=np.array([7,4,7,3,7,3,6,3,7])
x, y = np.meshgrid(x, y)
z= np.sin((x**2 + y**2)/4)
ax1.plot_surface(x,y,z)
#ax1.plot_wireframe(x,y,z)
ax1.set_xlabel("X")
ax1.set_ylabel('y')
ax1.set_zlabel('z')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
x=np.array([1,2,3,7,6,7,5])
y=np.array([7,4,7,3,6,3,7])
x, y = np.meshgrid(x, y)
z= np.sin((x**2 + y**2)/4)

x1=np.array([1,4,5,2,5,3,5])
y1=np.array([6,7,4,6,3,6,6])
x1,y1=np.meshgrid(x,y)
z1=np.sin((x1**2+y1**2))

ax1.scatter(x1,y1,z1,color='c')
ax1.scatter(x,y,z,color='r')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
x=[1,2,3,4,2,4,6,5,3,6]
y=[4,2,4,2,4,3,6,4,3,5]
#x,y=np.meshgrid(x,y)
z=[4,2,4,2,4,2,7,5,4,6]
dx=np.ones(10)
dy=np.ones(10)
dz=[4,3,4,3,3,6,6,4,6,6]
ax1.bar3d(x,y,z,dx,dy,dz)
```
