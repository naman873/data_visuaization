```python
import urllib
import matplotlib.dates as mdates
```

```python
##reading file from interenet
```

```python
def bytespdate2num(b):
    return mdates.datestr2num(b.decode('utf-8'))
```

```python
def graph_data():
    stock_price_url="https://pythonprogramming.net/yahoo_finance_replacement"
    source=urllib.request.urlopen(stock_price_url).read().decode()
    
    stock_data=[]
    split_source=source.split("\n")
    
    for line in split_source:
        if "Date" not in line:
            stock_data.append(line)
            
    Date,Open,High,Low,Close,Adjusted_close,Volume=np.loadtxt(stock_data,delimiter=",",unpack=True,
                                                             converters={0: bytespdate2num })
    
    plt.plot(Date,Close,label="price")
    plt.xlabel("Dates")
    plt.ylabel("Close")
    plt.legend()
    plt.show()
```

```python
a=graph_data()
```

```python
def graph_data_sub():
    
    plt.figure()
    
    ax1=plt.subplot2grid((1,1),(0,0))
    
    stock_price_url="https://pythonprogramming.net/yahoo_finance_replacement"
    source=urllib.request.urlopen(stock_price_url).read().decode()
    
    stock_data=[]
    split_source=source.split("\n")
    
    for line in split_source:
        if "Date" not in line:
            stock_data.append(line)
            
    Date,Open,High,Low,Close,Adjusted_close,Volume=np.loadtxt(stock_data,delimiter=",",unpack=True,
                                                             converters={0: bytespdate2num })
    
    ax1.plot(Date,Close,label="price")
    
    ax1.plot([],[],label="loss",color="r",alpha=.5,linewidth=.5)
    ax1.plot([],[],label="gain",color="g",alpha=.5,linewidth=.5)
    
    ax1.fill_between(Date,Close,where=(Close>Close[0]),facecolor='g',alpha=.5)
    ax1.fill_between(Date,Close,where=(Close<Close[0]),facecolor='r',alpha=.5)
    #ax1.fill_between(Date,Close[0],150,alpha=.3)
    
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
    
    ax1.xaxis.label.set_color('c')
    ax1.yaxis.label.set_color('c')
    ax1.yaxis.label.set_rotation(0)
    plt.grid(True) ## c='g',linestyle='-',linewidth=5
    
    plt.xlabel("Dates")
    plt.ylabel("Close")
    plt.legend()
    plt.subplots_adjust(left=0.09,bottom=0.20,right=0.70,top=0.90,wspace=0,hspace=0)  ## bottom can not be greater
                                                                                    ## then top
    plt.show()
```

```python
b=graph_data_sub()
```

```python
def graph_data_sub2():
    
    plt.figure()
    
    ax1=plt.subplot2grid((1,1),(0,0))
    
    stock_price_url="https://pythonprogramming.net/yahoo_finance_replacement"
    source=urllib.request.urlopen(stock_price_url).read().decode()
    
    stock_data=[]
    split_source=source.split("\n")
    
    for line in split_source:
        if "Date" not in line:
            stock_data.append(line)
            
    Date,Open,High,Low,Close,Adjusted_close,Volume=np.loadtxt(stock_data,delimiter=",",unpack=True,
                                                             converters={0: bytespdate2num })
    
    ax1.plot(Date,Close,label="price")
    
    ax1.plot([],[],label="loss",color="r",alpha=.5,linewidth=.5)
    ax1.plot([],[],label="gain",color="g",alpha=.5,linewidth=.5)
    
    ax1.axhline(Close[0],color='k',linewidth=5)       ### black line used for divide 
    
    ax1.fill_between(Date,Close,where=(Close>Close[0]),facecolor='g',alpha=.5)    
    ax1.fill_between(Date,Close,where=(Close<Close[0]),facecolor='r',alpha=.5)
    ax1.fill_between(Date,Close[0],150,alpha=.3)
    
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
    
    #ax1.xaxis.label.set_color('c')
    #ax1.yaxis.label.set_color('c')
    ax1.yaxis.label.set_rotation(0)
    plt.grid(True) ## c='g',linestyle='-',linewidth=5
    
    ax1.spines['left'].set_color('c')         ## left line
    ax1.spines['right'].set_visible(False)   ## right 
    ax1.spines['top'].set_visible(False)
    
    ax1.spines['left'].set_linewidth(5)
    
    ax1.tick_params(axis='x',colors='#f06215')
    ax1.tick_params(axis='y',colors="#f02999")
    
    plt.xlabel("Dates")
    plt.ylabel("Close")
    plt.legend()
    plt.subplots_adjust(left=0.09,bottom=0.20,right=0.70,top=0.90,wspace=0,hspace=0)  ## bottom can not be greater
                                                                                    ## then top
    plt.show()
```

```python
import matplotlib.ticker as mticker
from mpl_finance import candlestick_ohlc      ### open high low close
```

```python
def graph_data_sub3():
    
    plt.figure()
    
    ax1=plt.subplot2grid((1,1),(0,0))
    
    stock_price_url="https://pythonprogramming.net/yahoo_finance_replacement"
    source=urllib.request.urlopen(stock_price_url).read().decode()
            
    stock_data=[]
    split_source=source.split("\n")
    
    for line in split_source:
        if "Date" not in line:
            stock_data.append(line)
            
    Date,Open,High,Low,Close,Adjusted_close,Volume=np.loadtxt(stock_data,delimiter=",",unpack=True,
                                                             converters={0: bytespdate2num })
    
    
    x=0
    y=len(Date)
    ohlc=[]
    
    if x<y:
        append_me= Date[x] ,Open[x] ,High[x] ,Low[x] ,Close[x],Adjusted_close[x],Volume[x]
        ohlc.append(append_me)
        x+=1

    
    candlestick_ohlc(ax1,ohlc,width=.1,colorup='y',colordown='b')  ## when lines go up it is black and when lines goes down is become red
    
    
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
        
    
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    #ax1.xaxis.set_major_locator(mticker.MaxNL)
    
    plt.xlabel("Dates")
    plt.ylabel("Close")
    plt.legend()
    plt.subplots_adjust(left=0.09,bottom=0.20,right=0.70,top=0.90,wspace=0,hspace=0)  ## bottom can not be greater
                                                                                    ## then top
    plt.show()
```

```python
b=graph_data_sub3()
```

```python
def graph_data_sub4():
    
    plt.figure()
    
    ax1=plt.subplot2grid((1,1),(0,0))
    
    stock_price_url="https://pythonprogramming.net/yahoo_finance_replacement"
    source=urllib.request.urlopen(stock_price_url).read().decode()
            
    stock_data=[]
    split_source=source.split("\n")
    
    for line in split_source:
        if "Date" not in line:
            stock_data.append(line)
            
    Date,Open,High,Low,Close,Adjusted_close,Volume=np.loadtxt(stock_data,delimiter=",",unpack=True,
                                                             converters={0: bytespdate2num })
    
    
    x=0
    y=len(Date)
    ohlc=[]
    
    if x<y:
        append_me= Date[x] ,Open[x] ,High[x] ,Low[x] ,Close[x],Adjusted_close[x],Volume[x]
        ohlc.append(append_me)
        x+=1

    
    candlestick_ohlc(ax1,ohlc,width=.1,colorup='y',colordown='b')  ## when lines go up it is black and when lines goes down is become red
    
    
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
        
        
        
    ax1.annotate('Bad News',(Date[1],Low[1]),xytext=(0.4,0.4),textcoords='axes fraction',arrowprops=
                dict(facecolor='grey',color='gray'))   ## this is used for palcing text with arror 
                                                        # which is represenred by arrowprops 
    
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    #ax1.xaxis.set_major_locator(mticker.MaxNL)
    
    font_dict={'family':'serif',
              'color':'darkred',
              'size':15}
    
    ax1.text(Date[1],Low[1],"Placing_text",fontdict=font_dict)  ## this is used for placing text in graph 
    
    plt.xlabel("Dates")
    plt.ylabel("Close")
    #plt.legend()
    plt.subplots_adjust(left=0.09,bottom=0.20,right=0.70,top=0.90,wspace=0,hspace=0)  ## bottom can not be greater
                                                                                    ## then top
    plt.show()
```

```python
graph_data_sub4()
```
