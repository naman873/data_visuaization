```python
import random 
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
```

```python
def create_plot():
    
    xs=[]
    ys=[]
    
    for i in range(10):
        x=i
        y=random.randrange(10)
        xs.append(x)
        ys.append(y)
       
    return xs,ys
```

```python
fig=plt.figure()
ax1=fig.add_subplot(221)  ###nrows,ncolums,index  
ax2=fig.add_subplot(222)
ax3=fig.add_subplot(223)
ax4=fig.add_subplot(224)
x,y=create_plot()
ax1.grid(False)
ax1.plot(x,y)
x,y=create_plot()
ax2.plot(x,y)
x,y=create_plot()
ax3.plot(x,y)
x,y=create_plot()
ax1.grid(False)
ax4.plot(x,y)
```

```python
ax1=plt.subplot2grid((6,1),(0,0),rowspan=2,colspan=1)
ax2=plt.subplot2grid((6,1),(2,0),rowspan=2,colspan=1)
ax3=plt.subplot2grid((6,1),(4,0),rowspan=2,colspan=1)

x,y=create_plot()
ax1.plot(x,y)
x,y=create_plot()
ax2.plot(x,y)
x,y=create_plot()
ax3.plot(x,y)
```
