```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3]
y=[4,5,6]
```

```python
plt.figure()
plt.scatter(x,y,label="pro",c="red",marker="*",s=150) ## this s is marker size
plt.xlabel("X")
plt.ylabel("Y")
plt.title("interting \n Check it out")
plt.legend()
plt.show()
```

```python
##stack plot
```

```python
x=[1,2,3,4,5]
```

```python
sleeping=[10,9,8,7,9]
eating=[3,3,4,2,1]
working=[7,5,6,5,6]
playing=[6,5,5,5,6]
```

```python
plt.figure()
plt.plot([],[],color="m",label="sleeping",linewidth=5)
plt.plot([],[],color="c",label="eating",linewidth=5)
plt.plot([],[],color="r",label="working",linewidth=5)
plt.plot([],[],color="k",label="playing",linewidth=5)
plt.stackplot(x,sleeping,eating,working,playing,colors=['m','c','r','k']) ## we cant label in stackplot
plt.legend()
plt.show()
```

```python
##pie chart
```

```python
slices=[7,2,2,11]
activites=['sleeping','eating','working','playing']
plt.pie(slices,labels=activites,startangle=90,shadow=True,explode=(0,.1,0,0),autopct="%1.1f%%")
plt.legend()
plt.show()
```
